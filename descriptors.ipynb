Descriptors
__set__, __get__, __delete__

```python
class StringValidator:
    def __set_name__(self, owner, name):
        # print(f"self = {self}")
        # print(f"owner = {owner}")
        # print(f"name = {name}")
        self.name = name
    def __get__(self, instance, owner):
        # print("__get__ called")
        # print(f"self = {self}")
        # print(f"instance = {instance}")
        # print(f"owner = {owner}")
        if instance is None:
            return self
        return instance.__dict__.get(self.name, None)

    def __set__(self, instance, value):
        # print("__set__ called")
        # print(f"self = {self}")
        # print(f"instance = {instance}")
        # print(f"value = {value}")
        if not isinstance(value, str):
            raise TypeError("Invalid type")
        if not value:
            raise ValueError("Invalid Value")
        instance.__dict__[self.name] = value
```

```python
class Person:
    name = StringValidator()
    surname = StringValidator()
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
p = Person("Alice", "Poghosyan")
print(p.name, p.surname)
p.name = 7
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def get_name(self):
        return self.name
    
    def set_name(self, new):
        self.name = new

    name = property(get_name, set_name)

print(Person.name)
```

```python
from pydantic import BaseModel, Field

class Person(BaseModel):
    name: str = Field(default="aaaaaaaaaaaaaaaaaaaaa", min_length=8, max_length=20)
    surname: str
    age: int = Field(..., gt=8, lt=100)


p = Person(name="aaaaaaaa", surname="Poghosyan", age=7)
print(p)
```
