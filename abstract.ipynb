```python
class WithSlots:
    __slots__ = "name","age","initials"
    def __init__(self,name,age,initials):
        self.name=name
        self.age = age
        self.initials = initials

class WithNoSlots:
    def __init__(self,name,age):
        self.name=name
        self.age = age
```

```python
import time
import sys

t = time.perf_counter()
o = WithSlots("aaa",12,"A.A.")
t2 = time.perf_counter()
print(t2-t)

print(sys.getsizeof(o))
```

```python
import time
import sys

t = time.perf_counter()
o = WithNoSlots("aaa",12)
t2 = time.perf_counter()
print(t2-t)

print(sys.getsizeof(o))
```

ABC = Abstract Base Class

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):
    def __init__(self, make, model):
        self.make = make
        self.model = model

    @abstractmethod
    def drive(self):
        print("Dricing a car")

# a = Car()
Car.mro()
```

```python
class SuperCar(Car):
    def __init__(self, make, model):
        super().__init__(make, model)
    def drive(self):
        super().drive()
        # print("Driving a Supercar")

a  = SuperCar("Toyota", "Camry")
a.drive()
```
