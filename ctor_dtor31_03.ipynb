```python
class Person:
    def __init__(self, name, age): #Constructor-??????
        self.name = name
        self.age = age

    @property
    def name(self):  #Selector
        return self.__name 
    
    @name.setter
    def name(self, new): #Mutator
        if not isinstance(new, str):
            raise TypeError("Invlaid Type")
        if not new:
            raise ValueError("Invalid Value")
        self.__name = new

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, new):
        if not isinstance(new, int):
            raise TypeError("Invalid Type")
        if new < 0:
            raise ValueError("Invalid value")
        self.__age = new

    # def __del__(self): #Destructor
    #     print("Deleting object")
    #     del self

    def __iter__(self):
        yield self.name
        yield self.age

    


p1 = Person("Aren", 15)
# print(p1.name)
for i in p1:
    print(i)
```

```python
class List(list):
    def __iter__(self):
        for i in range(len(self) - 1, -1, -1):
            yield self[i]

a = list([1,2,3])
a
```

```python
b = List([1,2,3])
```

```python
for i in b:
    print(i, end=" ")
```

```python
class Person:
    def __new__(cls, *args, **kwargs):
        print("__new__ called")
        instance = super().__new__(cls)
        return instance
    
    def __init__(self, name, age):
        print("__init__ called")
        self.name = name
        self.age = age
```

```python
p3 = Person("Levon", 15)
```

```python
class Person:
    def __new__(cls, name, age):
        print("__new__ called")
        instance = super().__new__(cls)
        instance.get_age = lambda: age
        instance.name = name
        instance.age = age
        return instance
```

```python
p4 = Person("Narek", 15)
p4.__dict__
```

```python
p4.get_age()
```

```python
class Decorator:
    def __init__(self, func):
        self.f = func
    def __call__(self, *args, **kwargs):
        print("Before")
        x=self.f(*args, **kwargs)
        print("After")
        return x


@Decorator  #foo = Decorator(foo)
def foo():
    print("Foo function")
    return 5

print(foo())
```

```python
class Person:
    __slots__ = ["name", "age"]
    def __init__(self, name, age):
        self.name = name
        self.age = age

p2 = Person("Max", 15)
```

```python
p2.surname = "Iloyan"
```

```python
class Student(Person):
    __slots__ = Person.__slots__ + ["gpa"]
    def __init__(self, name, age, gpa):
        super().__init__(name, age)
        self.gpa = gpa
```

```python
s1 = Student("Alexander", 15, 1.0)
```
