```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        ...

class Dog(Animal):
    def make_sound(self):
        print("Woof")

class Cat(Animal):
    def make_sound(self):
        print("Myau")

class Parrot(Animal):
    def make_sound(self):
        print("lekonk")


d = Dog()
c = Cat()
p = Parrot()

def make_sound(ls):
    for i in ls:
        i.make_sound()

l = [d,c,p]
make_sound(l)

# c.make_sound()
# d.make_sound()
# p.make_sound()
```

```python
from functools import singledispatch

@singledispatch
def foo(a):
    print(f"Processing {a} with type {type(a)}")

@foo.register
def _(a:int):
    print("Call for int")
    return a ** 2

@foo.register
def _(a:float):
    print("Call for float")
    return a ** 2


print(foo(3))
print(foo(3.5))
```

```python
from abc import ABC, abstractmethod
class Person(ABC):
    @abstractmethod
    def foo(self):
        ...

class Student:
    def foo(self):
       ...

Person.register(Student)
print(isinstance(Student(), Person))
print(issubclass(Student, Person))
```

```python
class Student:
    school = "ACT"
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa

    @classmethod
    def change_school(cls, new_school):
        print(cls)
        cls.school = new_school


```

```python
s = Student("Ashot", 15, 3.7)
s.change_school("Politexnik")
print(Student.school)
print(s.school)
```
